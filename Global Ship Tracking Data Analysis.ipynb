Loading the Automatic Identification System (AIS) vessel_information.csv data for manipulation. 

```python
data2 = pd.read_csv(r"C:\Users\mawul\Downloads\vessel_information.csv")

data2 
```

```python
table_vessel = pd.DataFrame(data2)

table_vessel.head(5)
```

Counting and outputing the count of vessels in the vessel_information.csv data

```python
table_vessel_count = table_vessel['type'].nunique()
table_vessel_count
```

Checking if there are duplicate entry of vessels in the vessel_information.csv data

```python
table_vessel_reccur = table_vessel[table_vessel['type'] ==100]
table_vessel_reccur
```

Merging the  Automatic Identification System (AIS) transit_segments.csv and the the Automatic Identification System (AIS) vessel_information.csv on the "mmsi" column using outer join function.

```python
outer_merge = table_transit.merge(table_vessel, how = 'outer', on ='mmsi')
outer_merge
```

Using another method to merge the Automatic Identification System (AIS) transit_segments.csv and the the Automatic Identification System (AIS) vessel_information.csv on the "mmsi" column.

```python
indirect_inner_join = outer_merge.dropna()
indirect_inner_join
```

Using the inner join function to merg the  Automatic Identification System (AIS) transit_segments.csv and the the Automatic Identification System (AIS) vessel_information.csv on the "mmsi" column.

```python
direct_inner_join = table_transit.merge(table_vessel, how = 'inner')
direct_inner_join
```
